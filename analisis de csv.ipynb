## Respuestas
cuantas canciones son bailables usando una restriccion

```python
# generacion de pandita(tomo una columna)
dance= spotify["danceability"]

# Pregunta 1

# Parametro de restriccion
D=0.7

# total de canciones
C=dance.count() 
#C1=len(dance)
#numero de canciones bailables sum-sum
C_B= sum(dance>D)

#proporcion de canciones bailables en la lista

P_B=C_B/C

print("existen "+str(C_B) + " que son bailables de un total de  " + str(C)) 
print("El "+ str(round(P_B,4)*100) + " % de las canciones son bailables segun un parametro D = 0.5")
```

```python
# grafica de bailable , categorizar, histograma


# Metodo matplotli.pyplot , mejor conocido plt

plt.hist(dance,30, facecolor="yellow")
plt.xlabel("Bailabilidad [ua]")
plt.ylabel("Numero de canciones")
plt.title("Numero de canciones respecto a la bailabilidad")
plt.show()
```

```python
dance.plot(kind="hist", bins=30)
plt.xlabel("Bailabilidad [ua]")
plt.ylabel("Numero de canciones")
plt.title("Numero de canciones respecto a la bailabilidad")
plt.show()
plt.show()
```

## Tendencia de duracion

```python
duracion= spotify["duration_ms"]

duracion_min = duracion/60000

#print(str(round(duracion_min,2)))

plt.hist(duracion_min,40, facecolor="orange")
plt.xlabel("duracion [min]")
plt.ylabel("Numero de canciones")
plt.title("Tendencia de duracion")
plt.show()
```

## No tienen instrumentalizacion

```python
instrumentalizacion=spotify["instrumentalness"]

plt.hist(instrumentalizacion,40, facecolor="gray")
plt.xlabel("instrumentalizacion")
plt.ylabel("Numero de canciones")
plt.title("Instrumentalizacion")
plt.show()


# Parametro de restriccion
P_I=0.3

# total de canciones
I=instrumentalizacion.count() 

#numero de canciones sin instrumentalizacion sum-sum
CSI= sum(instrumentalizacion<P_I)

#proporcion de canciones SIN instrumentalizacion en la lista

P_CSI=CSI/I

print("En esta playlist hay " +str(CSI) + " canciones sin instrumentalizacion") 
print("Lo que representa el " +str(round(P_CSI,4)*100) + " % ")
```
